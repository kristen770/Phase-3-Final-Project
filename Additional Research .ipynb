<b>soft</b> - 50818 (Good)- Soft water is surface water that contains low concentrations of ions and in particular is low in ions of calcium and magnesium. Soft water naturally occurs where rainfall and the drainage basin of rivers are formed of hard, impervious and calcium-poor rocks. 

<b>salty</b> - 4856 (Salty) - Seawater is water from a sea or ocean. On average, seawater in the world's oceans has a salinity of approximately 3.5%, or 35 parts per thousand. This means that for every 1 litre (1000 mL) of seawater there are 35 grams of salts (mostly, but not entirely, sodium chloride) dissolved in it. 

<b>unknown</b> - 1876 (Unknown) - NA 

<b>milky</b> - 804 (milky)- milky or white water is likely due to trapped air or a build-up of dissolved particles in the water pipes

Slight discoloration is measured in Hazen units (HU) Impurities can be deeply colored as well, for instance dissolved organic compounds called tannins can result in dark brown colors, or algae floating in the water (particles) can impart a green color.

### Water Point Type Descriptions 
<details>
  <summary>Click to expand!</summary>
<b>communal standpipe</b> - 34625 - A standpipe is a freestanding pipe fitted with a tap which is installed outdoors to dispense water in areas which do not have a running water supply to the buildings. 

<b>hand pump</b> - 17488 - Hand pumps are manually operated pumps; they use human power and mechanical advantage to move fluids or air from one place to another.

<b>other</b> - 6380 - device unspecified

<b>improved spring</b> - 784 - A spring is developed by collecting the water. that flows out of a spring or seep in a pipe so. that the water can flow through a pipe to a. watering trough.
    
<b>cattle trough</b> - 116 - A watering trough (or artificial watering point) is a human-made or natural receptacle intended to provide drinking water to animals, livestock on farms or ranches or wild animals.

<b>dam</b> - 7 - A dam is a structure built across a river or stream to hold back water
